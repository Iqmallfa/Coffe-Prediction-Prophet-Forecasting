```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error

from prophet import Prophet
```

```python
#Import Dataset
df = pd.read_csv(io.BytesIO(uploaded['index.csv']))
print(df)
```

```python
#Analisis Dataset Coffe
prices = np.array(df.money)
dates = np.array(df.datetime)

fig, ax = plt.subplots(figsize=(18, 8))
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.plot(dates,prices)
ax.title.set_text("Price Per Date")
ax.set_xlabel("datetime")
ax.set_ylabel("Price")
```

```python
#Preparasi Data
df_a = df.loc[df['coffee_name'] == 'Americano'][['money', 'datetime']]
df_b = df.loc[df['coffee_name'] == 'Americano with Milk'][['money', 'datetime']]
df_c = df.loc[df['coffee_name'] == 'Cappuccino'][['money', 'datetime']]
df_d = df.loc[df['coffee_name'] == 'Cocoa'][['money', 'datetime']]
df_e = df.loc[df['coffee_name'] == 'Cortado'][['money', 'datetime']]
df_f = df.loc[df['coffee_name'] == 'Espresso'][['money', 'datetime']]
df_g = df.loc[df['coffee_name'] == 'Hot Chocolate'][['money', 'datetime']]
df_h = df.loc[df['coffee_name'] == 'Latte'][['money', 'datetime']]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
plt.style.use('fivethirtyeight')
```

```python
#Format Dataset Untuk Prophet Forecasting
rename_cols = {'datetime': 'ds', 'money': 'y'}
```

```python
df_a = df_a.reset_index(drop=True).rename(columns=rename_cols)
df_b = df_b.reset_index(drop=True).rename(columns=rename_cols)
df_c = df_c.reset_index(drop=True).rename(columns=rename_cols)
df_d = df_d.reset_index(drop=True).rename(columns=rename_cols)
df_e = df_e.reset_index(drop=True).rename(columns=rename_cols)
df_f = df_f.reset_index(drop=True).rename(columns=rename_cols)
df_g = df_g.reset_index(drop=True).rename(columns=rename_cols)
df_h = df_h.reset_index(drop=True).rename(columns=rename_cols)
```

```python
df_a
```

```python
model_a = Prophet()
model_b = Prophet()
model_c = Prophet()
model_d = Prophet()
model_e = Prophet()
model_f = Prophet()
model_g = Prophet()
model_h = Prophet()
```

```python
model_a.fit(df=df_a)
model_b.fit(df=df_b)
model_c.fit(df=df_c)
model_d.fit(df=df_d)
model_e.fit(df=df_e)
model_f.fit(df=df_f)
model_g.fit(df=df_g)
model_h.fit(df=df_h)
```

```python
# Create dataframe to store the predictions for 2018-2019
future_a = model_a.make_future_dataframe(periods=365)
future_b = model_b.make_future_dataframe(periods=365)
future_c = model_c.make_future_dataframe(periods=365)
future_d = model_d.make_future_dataframe(periods=365)
future_e = model_e.make_future_dataframe(periods=365)
future_f = model_f.make_future_dataframe(periods=365)
future_g = model_g.make_future_dataframe(periods=365)
future_h = model_h.make_future_dataframe(periods=365)
```

```python
predictions_a = model_a.predict(future_a)
predictions_b = model_b.predict(future_b)
predictions_c = model_c.predict(future_c)
predictions_d = model_d.predict(future_d)
predictions_e = model_e.predict(future_e)
predictions_f = model_f.predict(future_f)
predictions_g = model_g.predict(future_g)
predictions_h = model_h.predict(future_h)
```

```python
predictions_a
```

```python
with open('predictions_a.csv', 'w') as f:
    predictions_a.to_csv(f)
    predictions_b.to_csv(f)
    predictions_c.to_csv(f)
    predictions_d.to_csv(f)
    predictions_e.to_csv(f)
    predictions_f.to_csv(f)
    predictions_g.to_csv(f)
    predictions_h.to_csv(f)
```

```python
# Plot forecast
f, ax = plt.subplots(1)
plt.xticks(rotation=90)
plt.suptitle('Prediction Forecast')
fig = model_a.plot(predictions_a, ax=ax)
```
